```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('/Users/idilunlu/Desktop/data_filtering/lgg_tcga_pan_can_atlas_2018_clinical_data.tsv', sep='\t')

# Pie chart for gender distribution
if 'Sex' in data.columns:
    gender_counts = data['Sex'].value_counts()
    plt.figure(figsize=(10, 8))
    plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Gender')
    plt.axis('equal')
    plt.show()

# Categorizing age data into groups and generating pie chart
if 'Diagnosis Age' in data.columns:
    bins = [0, 18, 35, 55, 75, 100]
    labels = ['0-18 years', '19-35 years', '36-55 years', '56-75 years', '76+ years']
    data['age_group'] = pd.cut(data['Diagnosis Age'], bins=bins, labels=labels, right=False)
    age_group_counts = data['age_group'].value_counts()
    plt.figure(figsize=(10, 8))
    plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Age Groups')
    plt.axis('equal')
    plt.show()

# Pie chart for disease distribution
if 'Cancer Type Detailed' in data.columns:
    disease_counts = data['Cancer Type Detailed'].value_counts()
    plt.figure(figsize=(10, 8))
    plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Diseases')
    plt.axis('equal')
    plt.show()

# Pie chart for race distribution
if 'Race Category' in data.columns:
    race_counts = data['Race Category'].value_counts()
    plt.figure(figsize=(10, 8))
    plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of Race')
    plt.axis('equal')
    plt.show()
```
